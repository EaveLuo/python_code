```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + x

# 配置
matplotlib.rcParams['font.family'] = 'SimHei'  # 设置字体
matplotlib.rcParams['axes.unicode_minus'] = False  # 设置 负号 正常显示

plt.grid()

plt.title('f(x)图像')
plt.xlabel('x轴')
plt.ylabel('y轴')

# plt.ylim(0, 1)
# plt.xlim(0, 2*np.pi)

plt.plot(x, y, label='sinx')
plt.legend()
plt.show()
```

```python
import numpy as np
a = np.array([1, 0, 2, 0, 0, 1])
f = np.poly1d(a)
f1 = np.polyder(f, 1)
f2 = np.polyder(f, 2)
print("一阶导数：", f1(5))
print("二阶导数：", f2(5))
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np

df = pd.read_excel("./文件/studentscore.xlsx")
print(df)
print('-'*50)

df['总分'] = df.iloc[:, 2:].apply(lambda x:x.sum(),axis =1)
print(df)
print('-'*50)

df['均分'] = df.iloc[:, 2:-1].apply(lambda x:np.mean(x),axis =1)
print(df)
print('-'*50)

# 保存
df.to_csv('./文件/studentscore.csv')

print(df[df['均分']>90][['学号', '姓名']])
print('-'*50)

print(df[df['数学']>df['数学'].mean()][['学号', '姓名']])
print('-'*50)

print(df.sort_values(by=['总分'], ascending=False))
print('-'*50)


# 配置
matplotlib.rcParams['font.family'] = 'SimHei'  # 设置字体
matplotlib.rcParams['axes.unicode_minus'] = False  # 设置 负号 正常显示

plt.figure(figsize=(30, 50))
n = 1
for i in df.columns.tolist()[2:-2]:
    plt.subplot(5, 1, n)
    
    plt.title(i)
    plt.xlabel('x轴')
    plt.ylim(min(df[i]), max(df[i]))
    
    plt.bar(df['姓名'], df[i])
    
    n += 1
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np


df1 = pd.read_excel('./文件/meal_order_detail.xlsx')
df2 = pd.read_csv('./文件/meal_order_info.csv', encoding='gbk')
df3 = pd.read_excel('./文件/meal_users.xlsx')

print(f'meal_order_detail.xlsx记录条数{len(df1)}、列个数{len(df1.columns)}，列名称{df1.columns.tolist()}')
print(f'meal_order_info.csv记录条数{len(df2)}、列个数{len(df2.columns)}，列名称{df2.columns.tolist()}')
print(f'meal_users.xlsx记录条数{len(df3)}、列个数{len(df3.columns)}，列名称{df3.columns.tolist()}')
print('-'*50)


caidan = df1.groupby('dishes_name').agg({"dishes_name":"count"})
print(caidan)
plt.xticks(rotation=90)
plt.bar(caidan.index.tolist()[:10], [i[0] for i in caidan.values[:10]])
plt.show()
print('-'*50)

spid = df2.groupby('emp_id').agg({"emp_id":"count"})
print(spid)
plt.xticks(rotation=90)
plt.bar(spid.index.tolist()[:10], [i[0] for i in spid.values[:10]])
plt.show()
print('-'*50)

```
